```python
import pandas as pd
```

```python
dados = pd.read_csv('arquivos/aluguel.csv',sep = ';')
```

```python
dados.head()
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = ['Quitinete',
              'Casa',
              'Apartamento',
              'Casa de Condomínio',
              'Casa de Vila',
              ]
```

```python
dados_residenciais = dados[dados['Tipo'].isin(residencial)]
```

```python
dados_residenciais.head()
```

```python
dados_residenciais.info()
```

```python
dados_residenciais[dados_residenciais['Condominio'].isnull()]
```

```python
dados_residenciais.shape[0]
```

```python
dados_residenciais.index = range(dados_residenciais.shape[0])
dados_residenciais
```

```python
selecao = (dados_residenciais['Tipo'] == 'Apartamento') & (dados_residenciais['Condominio'].isnull())
selecao
```

```python
dados_residenciais[selecao]
```

```python
dados_residenciais = dados_residenciais[~selecao]
dados_residenciais
```

```python
dados_residenciais[dados_residenciais['Condominio'].isnull()].shape[0]
```

```python
dados_residenciais[dados_residenciais['Condominio'].isnull()]
```

```python
substituicao = {'Condominio': 0, 'IPTU': dados_residenciais.IPTU.mean()}
dados_residenciais = dados_residenciais.fillna(substituicao)
```

```python
dados_residenciais[dados_residenciais['Valor'].isnull()]
```

```python
selecao = dados_residenciais['Valor'].isnull()
dados_residenciais = dados_residenciais[~selecao]
```

```python
dados_residenciais.info()
```

```python
dados_residenciais.to_csv('arquivos/aluguel_residencial.csv',sep = ';', index = False)
```
