```python
import pandas as pd
import seaborn as sns
```

```python
dados = pd.read_csv('arquivos/aluguel_residencial.csv',sep =';')
```

```python
dados
```

```python
sns.boxplot(dados.Valor)
```

```python
q1 = dados['Valor'].quantile(.25)
q3 = dados['Valor'].quantile(.75)
iiq = q3 - q1
limite_superior = q3 + 1.5*iiq
limite_inferior = q1 - 1.5*iiq
```

```python
selecao = (dados['Valor'] > limite_inferior) & (dados['Valor'] < limite_superior)
dados_limpos = dados[selecao]
dados_limpos
```

```python
sns.boxplot(dados_limpos['Valor'])
```

```python
sns.distplot(dados_limpos['Valor'])
```
